# Version 1
only draw contours as they are and calculate the average char size per image

A: is for using extract_textarea function 
B: is for using the thresholded image directly

```python
def character_size(img):
    contours, hierarchy = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
    sizes = np.zeros(0)
    for contour in contours:
        (x,y,w,h) = cv2.boundingRect(contour)
        size = w*h
        if (size<500000):
            cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
            sizes = np.append(sizes, size)
    show_image(img)
    print("Max = ")
    if(len(sizes)) != 0:
        print(np.max(sizes))
        return np.mean(sizes)
    else:
        return 0

```

# Version 2
draw contours but ignore sizes < 10 && >100

```python
def character_size_2(img):
    contours, hierarchy = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
    sizes = np.zeros(0)
    for contour in contours:
        (x,y,w,h) = cv2.boundingRect(contour)
        size = w*h
        if (size > 100 ) and (size<10000) :
            cv2.rectangle(img, (x,y), (x+w,y+h), (0,(size / 10000) * 255,0), 2)
            sizes = np.append(sizes, size)
    show_image(img)
    print("Max = ")
    if(len(sizes)) != 0:
        print(np.max(sizes))
        return np.mean(sizes)
    else:
        return 0

```

```python
females_2 = []
img = cv2.imread("data/Females/F1.jpg")
thresh = threshold_image(img)
text = extract_textarea(thresh)
avg = character_size_2(text)
females_2.append(avg)
print(avg)


img = cv2.imread("data/Females/F10.jpg")
thresh = threshold_image(img)
text = extract_textarea(thresh)
avg = character_size_2(text)
females_2.append(avg)
print(avg)


img = cv2.imread("data/Females/F2.jpg")
thresh = threshold_image(img)
text = extract_textarea(thresh)
avg = character_size_2(text)
females_2.append(avg)
print(avg)



img = cv2.imread("data/Females/F5.jpg")
thresh = threshold_image(img)
text = extract_textarea(thresh)
avg = character_size_2(text)
females_2.append(avg)
print(avg)


img = cv2.imread("data/Females/F3.jpg")
thresh = threshold_image(img)
text = extract_textarea(thresh)
avg = character_size_2(text)
females_2.append(avg)
print(avg)
```
